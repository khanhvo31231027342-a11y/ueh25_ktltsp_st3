```python
import numpy as np
import pandas as pd
import sweetviz as sv
from autoviz.AutoViz_Class import AutoViz_Class
```

Sweetviz

```python
# 1. Tải dữ liệu Marketing Campaign
df = pd.read_csv("Data_Python_ST3/BT1/marketing_campaign.csv", sep = '\t')

# 2. Tạo báo cáo
# Lệnh này sẽ phân tích TẤT CẢ các biến (đơn biến) và tương quan (hai biến)
my_report = sv.analyze(df)

# 3. Hiển thị báo cáo dưới dạng file HTML
# Một file tên 'SweetViz_Report.html' sẽ được tạo, bạn có thể mở nó bằng trình duyệt
my_report.show_html('SweetViz_Report.html')
```

```python
df.head()
```

```python
# 2. Tạo báo cáo, chỉ định 'Response' là biến mục tiêu
# SweetViz sẽ tự động so sánh 'Income' vs 'Response', 'Age' vs 'Response', v.v.
# Đây chính là "phân tích hai biến" một cách tự động
report_with_target = sv.analyze(df, target_feat='Response')

# 3. Hiển thị báo cáo
report_with_target.show_html('SweetViz_Target_Report.html')
```

Autoviz

```python
# 2. Khởi tạo đối tượng AutoViz
AV = AutoViz_Class()

# 3. Chạy AutoViz
# Lệnh này sẽ tự động mở các cửa sổ biểu đồ (hoặc hiển thị trong Colab)
# Nó sẽ tự động thực hiện hàng loạt phân tích đơn biến và hai biến
dft = AV.AutoViz(filename="", sep=',', depVar='', dfte=df, header=0, verbose=2)
```
