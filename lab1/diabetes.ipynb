```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from scipy import stats
```

```python
diabetes = pd.read_csv("Data_Python_ST3/BT1/diabetes.csv")
df = diabetes.copy()
# Take a quick look at the data
print(df.dtypes)
print(df.shape)
df.head(10).style.background_gradient(cmap = "Purples_r")
```

```python
df.drop_duplicates(inplace=True)
df.isnull().sum()
```

```python
df.describe().T.style.background_gradient(cmap = "magma")
```

II. Xử lí và trực quan hóa dữ liệu

```python
fig, axes = plt.subplots(1, 3, figsize=(24, 7))
fig.suptitle('Phân tích trực quan các yếu tố liên quan đến bệnh tiểu đường', fontsize=18)
# Biểu đồ 1
sns.countplot(ax=axes[0], data=df, x='Outcome', palette='pastel', hue='Outcome', legend=False)
axes[0].set_title('1. Phân bố số lượng Bệnh nhân (Outcome)', fontsize=14)
axes[0].set_xlabel('Kết quả (0 = Không bệnh, 1 = Bệnh)', fontsize=12)
axes[0].set_ylabel('Số lượng (người)', fontsize=12)

# Biểu đồ 2
sns.boxplot(ax=axes[1], data=df, x='Outcome', y='Glucose', palette='Blues', hue='Outcome', legend=False)
axes[1].set_title('2. Phân bố Glucose theo Outcome', fontsize=14)
axes[1].set_xlabel('Kết quả', fontsize=12)
axes[1].set_ylabel('Nồng độ Glucose', fontsize=12)

# Biểu đồ 3
correlation_matrix = df.corr()
sns.heatmap(ax=axes[2], data=correlation_matrix, annot=True, cmap='coolwarm', fmt='.2f', annot_kws={"size": 8})
axes[2].set_title('3. Ma trận tương quan các yếu tố', fontsize=14)

plt.tight_layout()
plt.show()
```
